```python
print("Installation...")
!pip install groq ipywidgets -q
print("OK !\n")

from groq import Groq
import json
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

# Ta clé Groq
API_KEY = "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"
client = Groq(api_key=API_KEY)

# Variables globales
questions_data = []
reponse_widgets = []
output = widgets.Output()

# Interface principale
print("="*60)
print("🎓 GÉNÉRATEUR D'EXAMEN IA - Propulsé par Groq (Llama 3)")
print("="*60)

theme_input = widgets.Text(
    placeholder="Ex: Python, Révolution française, Intelligence Artificielle, Biologie...",
    description="Thème:",
    layout=widgets.Layout(width='70%')
)

generate_btn = widgets.Button(
    description="📝 Générer l'examen",
    button_style='primary',
    layout=widgets.Layout(width='200px')
)

def generer_examen(b):
    global questions_data, reponse_widgets

    with output:
        clear_output()
        theme = theme_input.value

        if not theme:
            print("❌ Veuillez entrer un thème !")
            return

        print(f"🤖 L'IA génère des questions sur '{theme}'...\n")

        prompt = f"""Génère 3 questions d'examen sur "{theme}".

        Les questions doivent être VARIÉES, PERTINENTES et de DIFFÉRENTS TYPES.

        Réponds UNIQUEMENT au format JSON, sans rien d'autre.

        Format JSON EXACT:
        [
            {{"texte": "QUESTION 1 (QCM): [question sur {theme}]", "options": "OPTIONS: A) ... B) ... C) ... D) ...", "reponse": "A"}},
            {{"texte": "QUESTION 2 (TEXTE): [question ouverte originale sur {theme}]", "options": "", "reponse": "[réponse courte]"}},
            {{"texte": "QUESTION 3 (QCM): [troisième question différente]", "options": "OPTIONS: A) ... B) ... C) ... D) ...", "reponse": "B"}}
        ]"""

        try:
            response = client.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=[{"role": "user", "content": prompt}],
                temperature=0.8
            )

            questions_data = json.loads(response.choices[0].message.content)
            afficher_questions()

        except Exception as e:
            print(f"❌ Erreur: {e}")

def afficher_questions():
    global reponse_widgets
    reponse_widgets = []

    clear_output(wait=True)

    print("\n" + "="*60)
    print("📋 VOTRE EXAMEN GÉNÉRÉ PAR L'IA")
    print("="*60 + "\n")

    for i, q in enumerate(questions_data):
        print(f"{q['texte']}")
        if q['options']:
            print(f"   {q['options']}")
        print()

        reponse = widgets.Text(
            placeholder="Écrivez votre réponse ici...",
            description=f"Q{i+1}:",
            layout=widgets.Layout(width='80%')
        )
        reponse_widgets.append(reponse)
        display(reponse)
        print()

    print("\n" + "-"*60)

    corriger_btn = widgets.Button(
        description="✅ CORRIGER MON EXAMEN",
        button_style='success',
        layout=widgets.Layout(width='300px')
    )
    corriger_btn.on_click(corriger_examen)
    display(corriger_btn)

def corriger_examen(b):
    global questions_data, reponse_widgets

    clear_output(wait=True)

    print("\n" + "="*60)
    print("📊 CORRECTION DE VOTRE EXAMEN")
    print("="*60 + "\n")

    score = 0
    details = []

    for i, q in enumerate(questions_data):
        bonne = q['reponse'].lower().strip()
        eleve = reponse_widgets[i].value.lower().strip()

        est_correct = False
        if bonne in eleve:
            est_correct = True
        elif bonne == "a" and eleve in ["a", "a)"]:
            est_correct = True
        elif bonne == "b" and eleve in ["b", "b)"]:
            est_correct = True
        elif bonne == "c" and eleve in ["c", "c)"]:
            est_correct = True
        elif bonne == "d" and eleve in ["d", "d)"]:
            est_correct = True
        elif len(eleve) > 10 and len(bonne) > 5:
            est_correct = True

        if est_correct:
            score += 1
            print(f"✅ Question {i+1}: CORRECT")
        else:
            print(f"❌ Question {i+1}: INCORRECT")
            print(f"   Réponse attendue: {q['reponse']}")
        print()

    print("-"*60)
    print(f"\n🎯 SCORE FINAL: {score}/{len(questions_data)}")

    if score == 3:
        print("🎉 PARFAIT ! Excellent travail !")
    elif score == 2:
        print("👍 TRÈS BIEN ! Continuez comme ça !")
    else:
        print("📚 BON COURAGE ! Révisez et réessayez !")

    print("\n" + "="*60)

    # Bouton pour recommencer
    nouveau_btn = widgets.Button(
        description="🔄 NOUVEL EXAMEN",
        button_style='warning',
        layout=widgets.Layout(width='200px')
    )
    nouveau_btn.on_click(lambda x: generer_examen(None))
    display(nouveau_btn)

# Affichage
display(theme_input, generate_btn, output)
generate_btn.on_click(generer_examen)

print("\n✅ Prêt ! Entrez un thème et cliquez sur 'Générer l'examen'")
```
